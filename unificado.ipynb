```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("C:\\Users\\kp6948\Documents\\thist\\corridas\\2731_Unificado_01_01_2023_20_03_2023.csv",encoding='latin-1')
```

```python

new_equipo = df['equipo'].str.split(pat=r" " ,expand=True  ,n=-1 ,regex=True ).fillna("")

new_equipo['concat'] = new_equipo[0] + "-" + new_equipo[1] + "-" + new_equipo[2] + "-" +new_equipo[3] + "-" +new_equipo[4]  

new_equipo['concat'].unique()
df['equipo'] = new_equipo['concat']
```

```python
new_0 = df['mensajeRPA'].str.split(pat=r"[^\\]+(.........[0-9]A)" ,expand=True  ,n=-1 ,regex=True ).fillna("")
new_0[1]





df['ordenmensajeRPA'] = new_0[1]
```

```python
new1 = df['crm'].str.split(pat=r"[^\\]+(.........[0-9]A)" ,expand=True ,n=-1 ,regex=True ).fillna("")
new1
df["ordencrm"]= new1[0]
```

```python
new2 = df['comentarios'].str.split(pat=r"[^\\]+(.........[0-9]A)" ,expand=True  ,n=-1 ,regex=True ).fillna("")
# new2
df["ordencomentarios"]= new2[1]
```

```python
df.head()
```

```python
df_2 = df[['ordencomentarios','ordenmensajeRPA']]

df_2['Col4'] = df[['ordencomentarios','ordenmensajeRPA']].apply(lambda x: ''.join(x.dropna()), axis=1)
```

```python
df_2
```

```python
df['ordenfinal'] = df_2['Col4'].str[-11:]
df['Origen'] = "Magento"
df['Fuente'] = "full_bd_unificado"
df['ComentarioDuplicadoECO'] = "No"
```

```python
df.reset_index(drop=True)
```

```python
a_index= ['increment_id',
 'fecha_de_pago',
 'status',
 'tipo',
 'guest',
 'region_factura',
 'oferta',
 'paquete',
 'mesesservicio',
 'sku',
 'precio_equipo',
 'precio_plan',
 'qty_ordered',
 'activacion',
 'total_pedido',
 'payment_method',
 'brand',
 'crm',
 'Origen',
 'Fuente',
 'ComentarioDuplicadoECO',
 'ordenmensajeRPA',
 'ordencrm',
 'ordencomentarios',
 'ordenfinal',
 'allow_esim']
df = df[a_index]

df.reindex(columns=[a_index])


df.to_csv(r'C:\\Users\\kp6948\Documents\\thist\\corridas\\corrida_2023_upsets_febrero_bd_clean.csv')
```
